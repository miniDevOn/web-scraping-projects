```python
base_url = 'https://understat.com/player/'
# Position or roles of football players
Roles = ['AMC','DR','DMR','MR','AMR','FWR','FWL','Sub','AML','FW','FWL','DML','DC','DL','DR','ML']

def playerShots(id):
    
    url = base_url + str(id)
    result = requests.get(url)
    soup = BeautifulSoup(result.content, 'lxml')
    scripts = soup.find_all('script')
    shots = scripts[3].string
    index_start = shots.index("('")+2
    index_end = shots.index("')")
    json_data = shots[index_start:index_end]
    json_data = json_data.encode('utf8').decode('unicode_escape') 
    data = json.loads(json_data)
    df = pd.DataFrame.from_dict(data)
    
    for r in range(len(df.index)):
        df.X[r]=pd.to_numeric(df.X[r])*100.0
        df.Y[r]=pd.to_numeric(df.Y[r])*100.0
        
    return df
```

```python
def matchesData(id):
    
    url = base_url + str(id)
    result = requests.get(url)
    soup = BeautifulSoup(result.content)
    scripts = soup.find_all('script')
    matches_data = scripts[4].string
    index_start = matches_data.index("('")+2
    index_end = matches_data.index("')")
    json_data = matches_data[index_start:index_end]
    json_data = json_data.encode('utf8').decode('unicode_escape')
    data = eval(json_data)
    df = pd.DataFrame.from_dict(data)
    
    return df
```

```python
kb10_matches= matchesData(2370)
kb10_matches
```

```python
def allPlayerData(id):
    
    url = base_url + str(id)
    result = requests.get(url)
    soup = BeautifulSoup(result.content)
    scripts = soup.find_all('script')
    g_data = scripts[1].string
    index_start = g_data.index("('")+2
    index_end = g_data.index("')")
    json_data = g_data[index_start:index_end]
    json_data = json_data.encode('utf8').decode('unicode_escape')
    data = eval(json_data)
    df = json_normalize(data)
    position_data = json_normalize(df['season'][0])
    
    return  position_data
```

```python
kb10_total= allPlayerData(2370)
kb10_total
```

```python
def radar_data(Pid: int, Role: str):
    
    url = base_url +str(Pid)
    try:
        request = requests.get(url)
    except ConnectionError as error:
        print('A connection problem', error)
        
    soup = BeautifulSoup(request.content)
    scripts  = soup.find_all('script')
    p_data = scripts[2].string
    index_start = p_data.index("('")+2
    index_end = p_data.index("')")
    json_data = p_data[index_start:index_end]
    json_data = json_data.encode('utf8').decode('unicode_escape')
    data = json.loads(json_data)
    data = pd.DataFrame.from_dict(data)
    min= ['min']
    max= ['max']
    avg= ['avg']
    
    for r in range(len(data)):
        d =json_normalize(data['{}'.format(Role)][r])
        min.append(d.iloc[0][0])
        max.append(d.iloc[0][1])
        avg.append(d.iloc[0][2])
        
    min_df=pd.DataFrame(min).T
    min_df.columns=['type','goals','xG','shots','Assists','xA','key_passes','xGChain','xGBuildup']
    max_df=pd.DataFrame(max).T
    max_df.columns=['type','goals','xG','shots','Assists','xA','key_passes','xGChain','xGBuildup']
    avg_df=pd.DataFrame(avg).T
    avg_df.columns=['type','goals','xG','shots','Assists','xA','key_passes','xGChain','xGBuildup']
    df = pd.concat([min_df,max_df,avg_df])
    df.set_index('type', inplace= True)
    return df
```

```python
kb10_radar= radar_data(2370, 'FW')
kb10_radar
```

```python
def league_data(league: str):
    
    base_url= 'https://understat.com/league/'
    url = base_url +league
    result = requests.get(url)
    soup = BeautifulSoup(result.content)
    scripts = soup.find_all('script')
    ps_data = scripts[3].string
    index_start = ps_data.index("('")+2
    index_end = ps_data.index("')")
    json_data = ps_data[index_start:index_end]
    json_data = json_data.encode('utf8').decode('unicode_escape')
    data = json.loads(json_data)
    data = pd.DataFrame.from_dict(data)
    
    return data
```

```python
def standings(league: str):
    
    base_url= 'https://understat.com/league/'
    url = base_url +league
    result = requests.get(url)
    soup = BeautifulSoup(result.content)
    scripts = soup.find_all('script')
    teams_data = scripts[2].string
    index_start = teams_data.index("('")+2
    index_end = teams_data.index("')")
    json_data = teams_data[index_start:index_end]
    json_data = json_data.encode('utf8').decode('unicode_escape')
    data = json.loads(json_data)
    data = pd.DataFrame.from_dict(data)
    data
    columns = ['team','M','W','D','L','G','Pts','xG','xGA','xPTS']
    team_df = pd.DataFrame()
    
    for c in range(len(data.columns)-1):
        teams_data = data.iloc[2][c]
        team = data.iloc[1][c]
        overall = json_normalize(teams_data)
        data_list = [team,len(overall),sum(overall['wins']),sum(overall['draws']),sum(overall['loses']),sum(overall['scored']),sum(overall['pts']),sum(overall['xG']),sum(overall['xGA']),sum(overall['xpts'])]
        team_df1 = pd.DataFrame(data_list).T
        team_df = pd.concat([team_df,team_df1])
        
    team_df.columns = columns
    return team_df
```

```python
serie_A_std= standings('Serie_A')
serie_A_std
```
